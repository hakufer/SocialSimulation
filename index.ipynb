# Ejercicio 1

```python
import itertools
from random import choice
import pandas as pd
```

```python
# Estrategias posibles
strategies = ['Rock', 'Paper', 'Scissors']
```

```python
payoff = {
    ('Rock', 'Paper'): (0, 1),
    ('Paper', 'Rock'): (1, 0),
    ('Rock', 'Scissors'): (1, 0),
    ('Scissors', 'Rock'): (0, 1),
    ('Paper', 'Scissors'): (0, 1),
    ('Scissors', 'Paper'): (1, 0),
    ('Rock', 'Rock'): (0, 0),
    ('Paper', 'Paper'): (0, 0),
    ('Scissors', 'Scissors'): (0, 0)
}
```

```python
# Definir los jugadores
names = ['Jim', 'Jane', 'Peter', 'Zoe']
society = [{'name': n, 'score': 0, 'strategy': None} for n in names]
```

```python
# Información adicional para seguimiento
print("¡Comienza el torneo de Piedra, Papel o Tijera!\n")
```

```python
rounds = 5  # Número de rondas para hacer el torneo más interesante
for round_num in range(1, rounds + 1):
    print(f"\n--- Ronda {round_num} ---\n")
    # Cada ronda implica enfrentamientos entre cada par de jugadores
    for player1, player2 in itertools.combinations(society, 2):
        # Cada jugador elige una estrategia aleatoriamente
        player1['strategy'] = choice(strategies)
        player2['strategy'] = choice(strategies)
        
        # Determinar el resultado del enfrentamiento
        result = payoff[player1['strategy'], player2['strategy']]
        
        # Actualizar el puntaje de cada jugador según el resultado
        player1['score'] += result[0]
        player2['score'] += result[1]
        
        # Mostrar el enfrentamiento y el resultado
        print(f"{player1['name']} juega {player1['strategy']} contra {player2['name']} que juega {player2['strategy']}.")
        if result[0] > result[1]:
            print(f"Resultado: {player1['name']} gana este enfrentamiento.")
        elif result[1] > result[0]:
            print(f"Resultado: {player2['name']} gana este enfrentamiento.")
        else:
            print("Resultado: Empate.")
```

```python
# Mostrar la tabla de puntajes después de cada ronda
socialResults = pd.DataFrame(society)
print("\nPuntajes tras la ronda:")
print(socialResults[['name', 'score']])
```

```python
# Determinar el puntaje más alto y quiénes son los líderes
max_score = socialResults['score'].max()
leaders = socialResults[socialResults['score'] == max_score]

# Verificar si hay empate o un líder claro
if len(leaders) > 1:
    print("\nActualmente hay un empate en el primer lugar entre:")
else:
    print("\nActualmente el líder es:")

# Mostrar los líderes actuales
print(leaders[['name', 'score']])
```

```python
print("\n=== Resultados Finales del Torneo ===")
socialResults = pd.DataFrame(society)
winnerScore = socialResults['score'].max()
winners = socialResults[socialResults['score'] == winnerScore]

if len(winners) > 1:
    print("\nEmpate final entre los siguientes jugadores:")
else:
    print("\nEl ganador del torneo es:")
print(winners[['name', 'score']])
```

# Ejercicio 3

```python
import random
```

```python
class Player:
    def __init__(self, name, score=0):
        self.name = name
        self.score = score
        self.strategy = None

    def get_strategy(self):
        # Elegir una estrategia aleatoria entre Piedra, Papel, Tijeras
        strategies = ['Rock', 'Paper', 'Scissors']
        self.strategy = random.choice(strategies)
        return self.strategy

    def increase_score(self, value):
        self.score += value
```

```python
payoff = {
    ('Rock', 'Rock'): (0, 0),
    ('Rock', 'Paper'): (0, 1),
    ('Rock', 'Scissors'): (1, 0),
    ('Paper', 'Rock'): (1, 0),
    ('Paper', 'Paper'): (0, 0),
    ('Paper', 'Scissors'): (0, 1),
    ('Scissors', 'Rock'): (0, 1),
    ('Scissors', 'Paper'): (1, 0),
    ('Scissors', 'Scissors'): (0, 0),
}
```

```python
class Game:
    def __init__(self, players):
        self.players = players

    def play_round(self, player1, player2):
        strategy1 = player1.get_strategy()
        strategy2 = player2.get_strategy()
        payoffs = payoff[(strategy1, strategy2)]
        player1.increase_score(payoffs[0])
        player2.increase_score(payoffs[1])
        print(f"{player1.name} ({strategy1}) vs {player2.name} ({strategy2}) -> Scores: {player1.score} - {player2.score}")

    def play_game(self, rounds=100):
        # Jugar 100 rondas
        for _ in range(rounds):
            # Cada ronda los jugadores se enfrentan entre ellos
            for i in range(len(self.players)):
                for j in range(i + 1, len(self.players)):
                    self.play_round(self.players[i], self.players[j])

    def declare_winner(self):
        # Encontrar al jugador con la puntuación más alta
        winner = max(self.players, key=lambda player: player.score)
        print(f"\nEl ganador es {winner.name} con un puntaje de {winner.score} puntos!")
```

```python
players = [Player("John"), Player("Mary"), Player("Alice"), Player("Bob")]
```

```python
game = Game(players)
```

```python
game.play_game(rounds=100)
```

```python
game.declare_winner()
```
